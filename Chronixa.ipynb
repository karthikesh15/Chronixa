```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import joblib

def generate_balanced_data(n_samples=2000):
    np.random.seed(42)
    n_per_class = n_samples // 2
    data = []

    # Generate Low Risk (Label 0): Healthy, younger, lower vitals
    for _ in range(n_per_class):
        data.append({
            'age': np.random.randint(20, 50), 'sex': np.random.randint(0, 2),
            'bmi': np.random.uniform(18, 25), 'chronic_hf': 0, 'diabetes': 0,
            'weight_gain_pct': np.random.uniform(0, 1.5), 'mean_sbp': np.random.uniform(110, 130),
            'mean_hr': np.random.uniform(60, 75), 'mean_spo2': np.random.uniform(97, 99),
            'hbA1c': np.random.uniform(4.0, 5.5), 'bnp': np.random.uniform(10, 80),
            'adherence_30': np.random.uniform(0.9, 1.0), 'ed_visits_90': 0, 'prior_admits': 0,
            'smoking': 0, 'alcohol': 0, 'diet_score': np.random.uniform(4.0, 5.0),
            'family_history': 0, 'socioecon': 1, 'med_count': np.random.randint(0, 2),
            'sleep_hrs': np.random.uniform(7, 9), 'recent_infection': 0, 'label': 0
        })

    # Generate High Risk (Label 1): Older, higher vitals, chronic indicators
    for _ in range(n_per_class):
        data.append({
            'age': np.random.randint(65, 90), 'sex': np.random.randint(0, 2),
            'bmi': np.random.uniform(29, 40), 'chronic_hf': 1, 'diabetes': 1,
            'weight_gain_pct': np.random.uniform(3, 7), 'mean_sbp': np.random.uniform(145, 190),
            'mean_hr': np.random.uniform(85, 110), 'mean_spo2': np.random.uniform(90, 94),
            'hbA1c': np.random.uniform(7.5, 12.0), 'bnp': np.random.uniform(300, 950),
            'adherence_30': np.random.uniform(0.3, 0.6), 'ed_visits_90': np.random.randint(1, 5),
            'prior_admits': np.random.randint(1, 4), 'smoking': 1, 'alcohol': 1,
            'diet_score': np.random.uniform(1.0, 2.5), 'family_history': 1, 'socioecon': 0,
            'med_count': np.random.randint(6, 15), 'sleep_hrs': np.random.uniform(4, 6),
            'recent_infection': np.random.randint(0, 2), 'label': 1
        })

    df = pd.DataFrame(data).sample(frac=1).reset_index(drop=True)
    return df

# Create and Train
df_synthetic = generate_balanced_data()
X = df_synthetic.drop('label', axis=1)
y = df_synthetic['label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = RandomForestClassifier(n_estimators=150, random_state=42)
model.fit(X_scaled, y)

# Save the new "Synthetic" versions
joblib.dump(model, 'chronic_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("Model trained on balanced synthetic data and saved!")
```
